```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
inventory = dict()
inventory["paracetamol"] = [50,1000]
inventory["painkiller"] = [70,2000]
```

```python
def additem(x,p,q):
    inventory[x] = [p,q]
```

```python
def updateitem(x,q):
    profit = inventory[x][0]*q
    inventory[x][1] -= q
    return profit
```

```python
def display_inv():
    print(inventory)
```

```python
#adding a new item 
additem("vix",90,600)
additem("xyz",100,900)


display_inv()
```

```python
x = np.linspace(1,10,10)
n = len(inventory)
sold = []

for i in range(n):
    sold.append(np.random.randint(1,6,10))
```

```python

#item wise profit print
fig , axs = plt.subplots(1,n)
keys = list(inventory.keys())
print(keys)
medicine_profits = dict()
profit = 0
day_wise_profit = np.zeros(10)
display_inv()
for i in range(n):
    axs[i].plot(x,sold[i])
    for j in range(10):
        temp = updateitem(keys[i],sold[i][j])
        day_wise_profit[j]+= temp
        profit += temp 
    medicine_profits[keys[i]] = profit
    profit = 0

#axs[n].plot(x,day_wise_profit)
print("profit")
print(medicine_profits)
print('current inventory')
display_inv()


total_profit = 0
for i in medicine_profits:
    total_profit += medicine_profits[i]
print("total profit ",total_profit)    
```
